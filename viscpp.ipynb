```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
N = 41
file = open('u.txt')
u = []
for line in file:
    a = line.split()
    u.append(a)
u = np.array(u).astype(np.float32)
file.close()
file = open('v.txt')
v = []
for line in file:
    a = line.split()
    v.append(a)
v = np.array(v).astype(np.float32)
file.close()
file = open('p.txt')
p = []
for line in file:
    a = line.split()
    p.append(a)
p = np.array(p).astype(np.float32)
file.close()
```

```python
Lx = 2.0
Ly = 2.0

#u = np.transpose(u)
#v = np.transpose(v)

dx = Lx/(N-1)
dy = Ly/(N-1)
x   = np.arange(0,Lx+dx,dx)
y   = np.arange(0,Ly+dy,dy)
#z   = np.arange(0,Lz+dz,dz)
x,y = np.meshgrid(x,y)
```

```python
fig = plt.figure(figsize=(11,7),dpi=100)
plt.contourf(x,y,u[:,:],alpha=0.5,cmap=cm.viridis)
plt.colorbar()
plt.contour(x,y,u[:,:],cmap=cm.viridis)
plt.streamplot(x,y,u,v)
plt.xlabel('x')
plt.ylabel('Y')
plt.savefig('U_VEL_CPP.png')
```

```python
fig = plt.figure(figsize=(11,7),dpi=100)
plt.contourf(x,y,v[:,:],alpha=0.5,cmap=cm.viridis)
plt.colorbar()
plt.contour(x,y,v[:,:],cmap=cm.viridis)
#plt.streamplot(x,y,u,v)
plt.xlabel('X')
plt.ylabel('Y')
plt.savefig('V_VEL_CPP.png')
```

```python
fig = plt.figure(figsize=(11,7),dpi=100)
plt.contourf(x,y,p[:,:],alpha=0.5,cmap=cm.viridis)
plt.colorbar()
plt.contour(x,y,p[:,:],cmap=cm.viridis)
#plt.streamplot(x,y,u,v)
plt.xlabel('X')
plt.ylabel('Y')
plt.savefig('P_VEL_CPP.png')
```

```python
fig = plt.figure(figsize=(10,6), dpi=100)
ax = plt.axes(projection='3d')

surf = ax.plot_surface(x, y, u,cmap='viridis', edgecolor='none')
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_title('Surface plot for x-component of Velocity', fontsize = 16)
plt.xlabel('x-ordinate', fontsize = 14)
plt.ylabel('y-ordinate', fontsize = 14)
plt.show()
plt.savefig('U_VEL_3d_CPP.png')
```

```python
fig = plt.figure(figsize=(10,6), dpi=100)
ax = plt.axes(projection='3d')

surf = ax.plot_surface(x, y, v,cmap='viridis', edgecolor='none')
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_title('Surface plot for y-component of Velocity', fontsize = 16)
plt.xlabel('x-ordinate', fontsize = 14)
plt.ylabel('y-ordinate', fontsize = 14)
plt.show()
plt.savefig('V_VEL_3d_CPP.png')
```

```python
fig = plt.figure(figsize=(10,6), dpi=100)
ax = plt.axes(projection='3d')

surf = ax.plot_surface(x, y, np.sqrt(np.power(u,2)+np.power(v,2)),cmap='viridis', edgecolor='none')
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_title('Surface plot of the Velocity Field U', fontsize = 16)
plt.xlabel('x-ordinate', fontsize = 14)
plt.ylabel('y-ordinate', fontsize = 14)
plt.show()
plt.savefig('NetU_VEL_3d_CPP.png')
```

```python
fig = plt.figure(figsize=(10,6), dpi=100)
ax = plt.axes(projection='3d')

surf = ax.plot_surface(x, y, p,cmap='viridis', edgecolor='none')
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set_title('Surface plot of the Pressure P', fontsize = 16)
plt.xlabel('x-ordinate', fontsize = 14)
plt.ylabel('y-ordinate', fontsize = 14)
plt.show()
plt.savefig('P_3d_CPP.png')
```
