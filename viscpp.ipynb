```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
N = 41
file = open('u.txt')
u = []
for line in file:
    a = line.split()
    u.append(a)
u = np.array(u).astype(np.float32)
file.close()
file = open('v.txt')
v = []
for line in file:
    a = line.split()
    v.append(a)
v = np.array(v).astype(np.float32)
file.close()
file = open('p.txt')
p = []
for line in file:
    a = line.split()
    p.append(a)
p = np.array(p).astype(np.float32)
file.close()
```

```python
Lx = 2.0
Ly = 2.0

#u = np.transpose(u)
#v = np.transpose(v)

dx = Lx/(N-1)
dy = Ly/(N-1)
x   = np.arange(0,Lx+dx,dx)
y   = np.arange(0,Ly+dy,dy)
#z   = np.arange(0,Lz+dz,dz)
x,y = np.meshgrid(x,y)
```

```python
fig = plt.figure(figsize=(11,7),dpi=100)
plt.contourf(x,y,u[:,:],alpha=0.5,cmap=cm.viridis)
plt.colorbar()
plt.contour(x,y,u[:,:],cmap=cm.viridis)
plt.streamplot(x,y,u,v)
plt.xlabel('x')
plt.ylabel('Y')
```

```python
fig = plt.figure(figsize=(11,7),dpi=100)
plt.contourf(x,y,v[:,:],alpha=0.5,cmap=cm.viridis)
plt.colorbar()
plt.contour(x,y,v[:,:],cmap=cm.viridis)
#plt.streamplot(x,y,u,v)
plt.xlabel('X')
plt.ylabel('Y')
```

```python
fig = plt.figure(figsize=(11,7),dpi=100)
plt.contourf(x,y,p[:,:],alpha=0.5,cmap=cm.viridis)
plt.colorbar()
plt.contour(x,y,p[:,:],cmap=cm.viridis)
#plt.streamplot(x,y,u,v)
plt.xlabel('X')
plt.ylabel('Y')
```
